```python
import tensorflow_datasets as tfds
import tensorflow as tf
```

```python
dataset, info = tfds.load("imdb_reviews/subwords8k",
                         with_info=True, as_supervised=True)
train_dataset, test_dataset = dataset["train"], dataset["test"]
```

```python
tokenizer = info.features["text"].encoder
```

```python
# Print sample shapes of first 4 sentences -> length is variable and not fixed
for index, i in enumerate(train_dataset.as_numpy_iterator()):
    print(i[0].shape, i[1].shape)
    if index == 3:
        break
```

```python
BUFFER_SIZE = 10000
BATCH_SIZE = 64

train_dataset = train_dataset.shuffle(BUFFER_SIZE)
padded_train_dataset = train_dataset.padded_batch(BATCH_SIZE,
                tf.compat.v1.data.get_output_shapes(train_dataset))
padded_test_dataset = test_dataset.padded_batch(BATCH_SIZE,
                tf.compat.v1.data.get_output_shapes(test_dataset))
```

```python
# Print sample shapes of first 4 sentences -> length is variable and not fixed
for index, i in enumerate(padded_train_dataset.as_numpy_iterator()):
    print(i[0].shape, i[1].shape)
    if index == 3:
        break
```

```python
padded_first_sample = list(padded_train_dataset.as_numpy_iterator())[0]
```

```python
print(padded_first_sample[0][3].shape)
print(padded_first_sample[0][3])
tokenizer.decode(padded_first_sample[0][3])
```

```python
# An error model because we specify input_length to be a specif number while each batch has different length
model = tf.keras.Sequential([
    # We cannot specify the input_length to be a specific number because each batch will have each different length
    tf.keras.layers.Embedding(input_dim = tokenizer.vocab_size+1, output_dim = 32, input_length = 16),
    tf.keras.layers.LSTM(32),
    tf.keras.layers.Dense(32, activation = 'relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])
model.summary()
model.compile(loss = 'binary_crossentropy', optimizer='adam', metrics=['accuracy'])
history = model.fit(x = padded_train_dataset, epochs=20, validation_data = padded_test_dataset)
```

```python
# An error model because we use Flatten when batch has different length -> Should change to GlobalAverage because
# this layer average out the last layer which is known before-hand
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(input_dim = tokenizer.vocab_size+1, output_dim = 32, input_length = None),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(32, activation = 'relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])
model.summary()
model.compile(loss = 'binary_crossentropy', optimizer='adam', metrics=['accuracy'])
history = model.fit(x = padded_train_dataset, epochs=20, validation_data = padded_test_dataset)
```

```python
import matplotlib.pyplot as plt


def plot_graphs(history, title):
    fig, axs = plt.subplots(1,2, figsize=(15,8))
    axs[0].plot(history.history['accuracy'], label='accuracy')
    axs[0].plot(history.history['val_accuracy'], label='val_accuracy')
    axs[1].plot(history.history['loss'], label='loss')
    axs[1].plot(history.history['val_loss'], label='val_loss')
    axs[0].set_title('Accuracy of '+ title)
    axs[1].set_title('Loss of '+ title)
    axs[0].set_xlabel("Epochs")
    axs[1].set_xlabel("Epochs")
    plt.legend()
    plt.show()
    
```

```python
plot_graphs(history, '1-Layer LSTM')
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(tokenizer.vocab_size, 64),
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(64)),
    tf.keras.layers.Dense(64, activation = 'relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])
model.summary()

model.compile(loss = 'binary_crossentropy', optimizer='adam', metrics=['accuracy'])
history = model.fit(x = train_dataset, epochs=20, validation_data = test_dataset)
```
